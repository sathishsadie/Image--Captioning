```python
images = coco_data['images']
annotations = coco_data['annotations']
```

```python
len(images)
```

```python
img_id_file = {img['id']:img['file_name'] for img in images}
```

```python
captions_data = {}
```

```python
# cnt = 500
for annot in annotations:
  id = annot['image_id']
  caption = annot['caption']
  file_name = img_id_file[id]
  if file_name in captions_data:
      pass
  else:
    captions_data[file_name] = caption
```
